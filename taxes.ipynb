```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
taxes = pd.read_csv('usgs_1800_2027.csv')
other = pd.read_csv('usgs_1800_2027(1).csv')
inflation = pd.read_csv('inflation_data.csv')

taxes = pd.concat([taxes, other], axis=1)
taxes[taxes.select_dtypes(include=['number']).columns] *= 10**9
start = 1913
end = 2021
taxes.Population /= 1000
taxes.Year /= 10**9

taxes = taxes[taxes.Year < 2020]

taxes.Total /= inflation.Amount
taxes.Wellfare /= inflation.Amount
taxes.Defense /= inflation.Amount
```

```python
plt.plot(taxes.Year, taxes.Total / taxes.Population)
plt.title('US Spending per Capita (Adjusted for inflaiton)')
plt.ylabel('Dollars ($)')
plt.xlabel('Year')
plt.show()

plt.plot(taxes.Year, taxes.Total / (10 ** 12))

plt.title('US Spending (Adjusted for inflaiton)')
plt.ylabel('Trillions of Dollars ($)')
plt.xlabel('Year')
plt.show()

plt.title('US Defense Spending per Capita (Adjusted for inflaiton)')
plt.plot(taxes.Year, taxes.Defense / taxes.Population)
plt.ylabel('Dollars ($)')
plt.xlabel('Year')
plt.show()

plt.title('US Defense Spending (Adjusted for inflaiton)')
plt.plot(taxes.Year, taxes.Defense / (10 ** 12))
plt.ylabel('Trillions of Dollars ($)')
plt.xlabel('Year')
plt.show()

plt.title('US Wellfare Spending per Capita (Adjusted for inflaiton)')
plt.plot(taxes.Year, taxes.Wellfare / taxes.Population)
plt.ylabel('Dollars ($)')
plt.xlabel('Year')
plt.show()

plt.title('US Wellfare Spending (Adjusted for inflaiton)')
plt.plot(taxes.Year, taxes.Wellfare / (10 ** 12))
plt.ylabel('Trillions of Dollars ($)')
plt.xlabel('Year')
plt.show()
```
